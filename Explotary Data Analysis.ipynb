```python
import time 
start= time.time()

vendor_sales_summary = pd.read_sql_query("""
WITH FreightSummary AS (
    SELECT
        VendorNumber,
        SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
),

PurchaseSummary AS (
    SELECT
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description,
             p.PurchasePrice, pp.Price, pp.Volume
),

SalesSummary AS (
    SELECT
        VendorNo,
        Brand,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
)

SELECT
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps.VendorNumber = ss.VendorNo
   AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC;
""", conn)
end = time.time()
timetaken = (end-start) / 60 
print(f"total time taken {timetaken} minutes")
```

```python
vendor_sales_summary["Volume"] = vendor_sales_summary["Volume"].astype("float") 
```

```python
#filling null values with 0 
vendor_sales_summary.fillna(0 , inplace=True)
```

```python
vendor_sales_summary["VendorName"]
```

```python
#removing white space from the vendor name column
vendor_sales_summary["VendorName"] = vendor_sales_summary["VendorName"].str.strip()
```

```python
vendor_sales_summary["VendorName"].unique()
```

```python
#calculating gross profit using total sales column and total purchase dollar column
vendor_sales_summary["GrossProfit"] = vendor_sales_summary["TotalSalesDollars"] - vendor_sales_summary["TotalPurchaseDollars"]
```

```python
vendor_sales_summary["GrossProfit"].min()
```

```python
#creating the profit margin column
vendor_sales_summary["ProfitMargin"] = (vendor_sales_summary["GrossProfit"] / vendor_sales_summary["TotalSalesDollars"]) * 100
```

```python
#Creating the column for the Stock Turn over
vendor_sales_summary["StockTurnOver"] = vendor_sales_summary["TotalSalesQuantity"] /vendor_sales_summary["TotalPurchaseQuantity"]
```

```python
#calculating Sale to Purchase Ratio
vendor_sales_summary["SaleToPurchaseRatio"] = vendor_sales_summary["TotalSalesDollars"] / vendor_sales_summary["TotalPurchaseDollars"]
```

```python
# creating the table , in this step we have just given the column
#name along with the data types

cursor.execute("""CREATE TABLE vendor_sales_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(100),
    PurchasePrice DECIMAL(10,2),
    ActualPrice DECIMAL(10,2),
    Volume,
    TotalPurchaseQuantity INT,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity INT,
    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    FreightCost DECIMAL(15,2),
    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalesToPurchaseRatio DECIMAL(15,2),
    PRIMARY KEY (VendorNumber, Brand)
);
""")

        

    
    
    
```

**DATA ANALYSIS**
 - identifying the performance
 - identifing the promotions

```python
brand_performance = df.groupby("Description").agg({
    'TotalSalesDollars':'sum',
    'ProfitMargin':'sum'}).reset_index()
```

```python
low_sales_thres = brand_performance["TotalSalesDollars"].quantile(0.15)
high_margin_thres = brand_performance["ProfitMargin"].quantile(0.85)
```

```python
# filtering those brands with low sales but high profit margin

target_brand = brand_performance[
    (brand_performance["TotalSalesDollars"] <= low_sales_thres) & 
    (brand_performance["ProfitMargin"] >= high_margin_thres)
]

print("Brands with the low sales but high profit ")

display(target_brand.sort_values("TotalSalesDollars"))
    
```

```python
import matplotlib.pyplot as plt

# Scatter plot of all brands
plt.figure(figsize=(10,6))
plt.scatter(
    brand_performance["TotalSalesDollars"],
    brand_performance["ProfitMargin"],
    alpha=0.7, label="All Brands"
)

# Highlight the target brands (low sales, high margin)
plt.scatter(
    target_brand["TotalSalesDollars"],
    target_brand["ProfitMargin"],
    color="red", s=100, label="Low Sales & High Profit"
)

# Draw threshold lines
plt.axvline(x=low_sales_thres, color="blue", linestyle="--", label=f"Low Sales â‰¤ {low_sales_thres}")
plt.axhline(y=high_margin_thres, color="green", linestyle="--", label=f"High Profit â‰¥ {high_margin_thres}")

# Labels and title
plt.xlabel("Total Sales Dollars")
plt.ylabel("Profit Margin")
plt.title("Brands with Low Sales but High Profit Margin")
plt.legend()
plt.grid(True)
plt.show()
```
